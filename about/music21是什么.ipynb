# Music21是什么
[原文链接](https://www.music21.org/music21docs/about/what.html)

Music21是一款基于Python的计算机乐理学套件。

通过Music21，可以用计算机解答乐理学所提出的问题，研究大规模的音乐数据集，生成音乐示例，教学基础音乐理论，编辑乐谱，研究音乐和大脑，以及作曲。

Music21的方针之一是“听得更快”（Listen faster）。使用本套件，你可以从一段音乐或一部音乐总集中发现有趣的瞬间，并获得一个总体的感受。我们希望：在计算机的协助下，你可以花更多时间为娱乐而听音乐与演奏音乐，用更少时间为工作而听音乐。

这套系统从2008年开始开发，至今仍在生长与扩展。Music21复用了许多其他软件系统用过的手段和惯例。本项目的作者和背景，参见[作者，鸣谢，贡献者与用户协议](作者，鸣谢，贡献者与用户协议.md)

Music21的21来自MIT最初孵化时的本项目。在MIT，所有与音乐与数学相关的课程，和一些其他人文学科相关的院系一道，被编号为21。MIT的音乐部门，和Harvard, Smith, 以及Mount Holyoke Colleges一起将这一套工具从最初的根基发展成成熟的系统。

## 迅速得到一个解决方案
music21向通用的，易于理解的“Python”编程语言加入了音乐领域的专用工具。安装music21，然后输入`python3`（或者，更好的选择是`ipython`），然后输入以下代码来载入music21：


```python
from music21 import *
```

然后数以千计的音乐工具等着你来使用。例如，想显示一个音符？请输入：


```python
n = note.Note("D#3")
n.duration.type = 'half'
n.show()
```

想要一串音符？更简单

```python
littleMelody = converter.parse("tinynotation: 3/4 c4 d8 f g16 a g f#")
littleMelody.show()
```

想听到这段旋律？同样简单！

```python
littleMelody.show('midi')
```

想用矩阵的方式查看Schoenberg的第四弦乐四重奏？

```python
print(serial.rowToMatrix([2, 1, 9, 10, 5, 3, 4, 0, 8, 7, 6, 11]))
```

用图表的方式展示一段14世纪作品中不同音高的出现频率

```python
dicant = corpus.parse('trecento/Fava_Dicant_nunc_iudei')
dicant.plot('histogram', 'pitch')
```

```python
dicant.show()
```

这个示例，以及以下的许多示例，来自music21内置的乐谱库。乐谱库包含几千段音乐作品让你迅速上手。我们把“自带电池”作为一项核心原则。例如，所有的巴赫赞美诗都收录在乐谱库中。向BWV295中的每个音符添加德语音名：

```python
bwv295 = corpus.parse('bach/bwv295')
for thisNote in bwv295.recurse().notes:
  thisNote.addLyric(thisNote.pitch.german)
bwv295.show()
```

为巴赫的所有3/4拍的作品建立一个摘要索引（这里只查看前25个）

```python
catalog = stream.Opus()
for work in corpus.chorales.Iterator(1, 26):
    firstTimeSignature = work.parts[0].measure(1).getTimeSignatures()[0]
    if firstTimeSignature.ratioString == '3/4':
        incipit = work.measures(0,2)
        catalog.insert(0, incipit.implode())

catalog.show()
```

本模块还包含高级的分析工具。想知道一个作品的节奏型有多么不稳定？用Ani Patel的nPVI函数：

```python
s = corpus.parse('AlhambraReel')
analysis.patel.nPVI(s.flatten())
```
